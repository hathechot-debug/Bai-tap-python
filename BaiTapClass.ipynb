#### Thông báo M có phải gốc tọa độ không hoặc có nằm trên trục tung hoặc trục hoành không, khoảng cách từ M tới gốc tọa độ là bao nhiêu (lấy 2 chữ số phần thập phân) 

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def goctoado(self) -> bool4:
        return self.x == 0 and self.y == 0
    def truchoanh(self):
        return self.x == 0 and self.y != 0
    def tructung(self):
        return self.x != 0 and self.y == 0
    def khoangcach(self, other):
        return ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
    def __str__(self):
        return (f"{self.x}, {self.y}")
    
x = int(input())
y = int(input())

A = Point(x, y)
O = Point(0 , 0)

print(A)
print(A.goctoado())
print(A.truchoanh())
print(f"{A.khoangcach(O):.2f}")
```

```python
class ShoppingCart:
    def __init__(self):
        self.products = []

    def add_product(self, name, price):
        product = {
            "name": name,
            "price": price
        }
        self.products.append(product)

    def remove_product(self, name):
        for product in self.products:
            if product["name"] == name:
                self.products.remove(product)
                return True
        return False

    def is_empty(self):
        return len(self.products) == 0

    def total_price(self):
        total = 0
        for product in self.products:
            total += product["price"]
        return total

    def show_products(self):
        if self.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            print("Danh sách sản phẩm trong giỏ:")
            for product in self.products:
                print("Tên:", product["name"], "- Giá:", product["price"])

    def clear_cart(self):
        self.products.clear()


cart = ShoppingCart()

while True:
    print("\n===== MENU SHOPPING CART =====")
    print("2. Thêm sản phẩm mới")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng có rỗng không")
    print("5. Tính tổng tiền giỏ hàng")
    print("6. Hiển thị toàn bộ sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Nhập lựa chọn từ 2 đến 7: ")

    if choice == "2":
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        cart.add_product(name, price)
        print("Đã thêm sản phẩm vào giỏ hàng.")

    elif choice == "3":
        name = input("Nhập tên sản phẩm cần xóa: ")
        if cart.remove_product(name):
            print("Đã xóa sản phẩm.")
        else:
            print("Không tìm thấy sản phẩm.")

    elif choice == "4":
        if cart.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            print("Giỏ hàng không rỗng.")

    elif choice == "5":
        print("Tổng tiền:", cart.total_price())

    elif choice == "6":
        cart.show_products()

    elif choice == "7":
        cart.clear_cart()
        print("Đã xóa toàn bộ giỏ hàng.")

    else:
        print("Lựa chọn không hợp lệ.")

    thoat = input("Có muốn thoát khỏi chương trình không? ")
    if thoat == "yes":
        break
```

### Yêu cầu
### Xây dựng lớp Rectangle(w, h) với các phương thức:
#### area() trả về diện tích.
#### perimeter() trả về chu vi.
#### scale(k) phóng to hình chữ nhật theo hệ số k (nhân cả w và h), không trả về gì.
### Input
#### w h k (số thực dương)
### Output
#### area perimeter sau khi phóng to (làm tròn 2 chữ số thập phân)

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w = self.w * k
        self.h = self.h * k


w, h, k = map(float, input().split())

rect = Rectangle(w, h)

rect.scale(k)

print(f"{rect.area():.2f}")
print(f"{rect.perimeter():.2f}")
```

```python
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")

        if den < 0:
            num = -num
            den = -den

        gcd = math.gcd(num, den)
        self.num = num // gcd
        self.den = den // gcd

    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)

    def div(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)

    def __str__(self):
        return f"{self.num}/{self.den}"


a, b, op, c, d = input().split()

a = int(a)
b = int(b)
c = int(c)
d = int(d)

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == "+":
    result = f1.add(f2)
elif op == "-":
    result = f1.sub(f2)
elif op == "*":
    result = f1.mul(f2)
elif op == "/":
    result = f1.div(f2)

print(result)
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, x):
        self.balance += x
    
    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

owner, balance = input().split()
balance = float(balance)
account = BankAccount(owner, balance)
n = int(input())
for i in range(n):
    command = input().split()

    if command[0] == "DEPOSIT":
        x = float(command[1])
        account.deposit(x)

    elif command[0] == "WITHDRAW":
        x = float(command[1])
        account.withdraw(x)

print(f"{account.balance:.2f}")
```
