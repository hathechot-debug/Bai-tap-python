#### Xây dựng lớp điểm với hai thuộc tính x, y cùng các phương thức kiểm tra xem có phải gốc tọa dộ không, có nằm trên trục tung không, có nằm trên trục hoành không, tính khoảng cách từ điểm đó tới một điểm khác
### Đầu vào:
#### Nhập vào tọa độ một điểm M bất kỳ
### Đầu ra:
#### Thông báo M có phải gốc tọa độ không hoặc có nằm trên trục tung hoặc trục hoành không, khoảng cách từ M tới gốc tọa độ là bao nhiêu (lấy 2 chữ số phần thập phân) 

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def goctoado(self) -> bool4:
        return self.x == 0 and self.y == 0
    def truchoanh(self):
        return self.x == 0 and self.y != 0
    def tructung(self):
        return self.x != 0 and self.y == 0
    def khoangcach(self, other):
        return ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
    def __str__(self):
        return (f"{self.x}, {self.y}")
    
x = int(input())
y = int(input())

p = Point(x, y)

print(p)
print(p.goctoado())
print(p.truchoanh())
print(p.khoangcach(Point(0, 0)))
```
