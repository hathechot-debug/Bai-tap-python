### W5A16
#### Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x


```python
def is_armstrong(x):
    if x < 0:
        return False

    s = str(x)
    n = len(s)

    total = 0
    for digit in s:
        total += int(digit) ** n

    return total == x


x = int(input())
print(is_armstrong(x))
```

### W5A17
### Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x


```python
def so_chinh_phuong(n):
    if n < 0:
        return False
    
    i = 0
    while i * i <= n:
        if i * i == n:
            return True
        i += 1
    return False

n = int(input())
print (so_chinh_phuong(n))
```

### W5A18
#### VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_palindrome_number(x):
    if x < 0:
        return False

    original = x
    so_doi_xung = 0

    while x > 0:
        digit = x % 10
        so_doi_xung = so_doi_xung * 10 + digit
        x //= 10

    return original == so_doi_xung


x = int(input("Nhập x: "))
print(is_palindrome_number(x))
```

### W5A19
#### VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_ky_dieu(n):
    if n <= 0:
        return False
    
    while n >= 10:
        total = 0
        while n > 0:
            digit = n % 10
            total += digit
            n //= 10
        n = total

    return n == 1

n = int(input())
print (so_ky_dieu(n))
```

### W5A22
#### Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

```python
def sum_halves(s):
    first = int(s[:3])
    second = int(s[3:])
    return first + second


s = input()
print(sum_halves(s))
```

### W5A31
####  Viết hàm tính tổng bình phương các số từ 1 đến n.


```python
def tong_binh_phuong(n):
    tong = 0
    for x in range (1, n + 1):
        x = x ** 2
        tong += x
    return tong

n = int(input())
print (tong_binh_phuong(n))
```
