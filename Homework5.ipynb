### W5A16
#### Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x


```python
def is_armstrong(x):
    if x < 0:
        return False

    s = str(x)
    n = len(s)

    total = 0
    for digit in s:
        total += int(digit) ** n

    return total == x


x = int(input())
print(is_armstrong(x))
```

### W5A17
### Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x


```python
def so_chinh_phuong(n):
    if n < 0:
        return False
    
    i = 0
    while i * i <= n:
        if i * i == n:
            return True
        i += 1
    return False

n = int(input())
print (so_chinh_phuong(n))
```

### W5A18
#### VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_palindrome_number(x):
    if x < 0:
        return False

    original = x
    so_doi_xung = 0

    while x > 0:
        digit = x % 10
        so_doi_xung = so_doi_xung * 10 + digit
        x //= 10

    return original == so_doi_xung


x = int(input("Nhập x: "))
print(is_palindrome_number(x))
```

### W5A19
#### VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_ky_dieu(n):
    if n <= 0:
        return False
    
    while n >= 10:
        total = 0
        while n > 0:
            digit = n % 10
            total += digit
            n //= 10
        n = total

    return n == 1

n = int(input())
print (so_ky_dieu(n))
```

### W5A27
#### Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)


```python
def dem_chu_so_khac_0(lst):
    dem = 0

    for n in lst:
        n = abs(n)

        if n == 0:
            continue

        while n > 0:
            i = n % 10
            if i != 0:
                dem += 1
            n //= 10

    return dem


lst = list(map(int, input().split()))
print(dem_chu_so_khac_0(lst))
```

### W5A28
#### Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math

def tinh_goc_ABC(x1, y1, x2, y2, x3, y3):
    # Vector BA = A - B
    BA_x = x1 - x2
    BA_y = y1 - y2

    # Vector BC = C - B
    BC_x = x3 - x2
    BC_y = y3 - y2

    # Tích vô hướng BA . BC
    dot = BA_x * BC_x + BA_y * BC_y

    # Độ dài vector BA và BC
    len_BA = math.sqrt(BA_x**2 + BA_y**2)
    len_BC = math.sqrt(BC_x**2 + BC_y**2)

    # Tính cos của góc ABC
    cos_angle = dot / (len_BA * len_BC)

    # Tránh lỗi do sai số máy tính
    cos_angle = max(-1, min(1, cos_angle))

    # Đổi từ radian sang độ
    angle = math.degrees(math.acos(cos_angle))

    return angle

x1, y1 = map(int, input().split())
x2, y2 = map(int, input().split())
x3, y3 = map(int, input().split())
print (tinh_goc_ABC(x1, y1, x2, y2, x3, y3))
```

### W5A30
#### Viết hàm tính hệ số nhị thức C(n, k).


```python
def he_so_nhi_thuc(n, k):
    if k < 0 or k > n:
        return 0

    # tính n!
    giai_thua_n = 1
    for i in range(1, n + 1):
        giai_thua_n *= i

    # tính k!
    giai_thua_k = 1
    for i in range(1, k + 1):
        giai_thua_k *= i

    # tính (n-k)!
    giai_thua_hieu = 1
    for i in range(1, n - k + 1):
        giai_thua_hieu *= i

    return giai_thua_n // (giai_thua_k * giai_thua_hieu)


n = int(input())
k = int(input())

print(he_so_nhi_thuc(n, k))
```

### W5A33
#### Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Bảng giá theo loại xe
    if car_type == "4cho":
        open_price = 12000
        price_per_km = 15000
    elif car_type == "7cho":
        open_price = 15000
        price_per_km = 18000
    else:
        return -1   # loại xe không hợp lệ

    # Kiểm tra dữ liệu không hợp lệ
    if start_hour < 0 or start_hour > 23 or km < 0 or wait_minutes < 0:
        return -1

    # Tiền mở cửa + tiền km
    total = open_price + km * price_per_km

    # Tiền chờ
    wait_fee = wait_minutes * 1000
    total += wait_fee

    # Phụ phí ban đêm: từ 22h đến trước 6h
    if start_hour >= 22 or start_hour < 6:
        total = total * 1.2

    # Giảm giá nếu đi trên 20km
    if km > 20:
        total = total * 0.9

    return int(total)

car_type = input()
start_hour = int(input())
km = int(input())
wait_minutes = int(input())
print (smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

### W5A34
#### Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau Nếu n ≤ 0 trả về 0.

```python
def chu_so_khac_nhau(x):
    s = str(x)

    for i in range(len(s)):
        for j in range(i + 1, len(s)):
            if s[i] == s[j]:
                return False

    return True


def sum_1_to_n(n):
    if n <= 0:
        return 0

    tong = 0

    for x in range(1, n + 1):
        if chu_so_khac_nhau(x):
            tong += x

    return tong

x = int(input())
print (sum_1_to_n(x))
```
