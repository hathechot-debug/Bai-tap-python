### Bài 4:
#### Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 


```python
h = int(input())

for i in range(1, h + 1):
    print(" " * (h - i), end="")
    
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")
```

### Bài 9:
### Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 

```python
a = int(input())
i = 1
while i * i < a:
    i += 1
if i * i == a:
    print (f"{a} la so chinh phuong")
else:
    print (f"{a} khong la so chinh phuong")
```

### Bài 12:
#### Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 


```python
n = int(input())

s_k = 0
k = 0
while s_k + (k + 1) < n:
    k += 1
    s_k += k

print(f"Số k lớn nhất để S(k) < {n} là: {k}")
print(f"Tổng S({k}) = {s_k}")
```

### Bài 14:
#### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
max_val = None
min_val = None
while True:
    n = int(input())
    if n == -1:
        break
    
    if max_val is None:
        max_val = n
        min_val = n
    else:
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n

if max_val is not None:
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
else:
    print("Bạn chưa nhập số nào hợp lệ ngoài số -1.")
```

### Bài 15:
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều cách nhất có thể, ko dùng mảng. 


```python
n = int(input())
count = 0
if n == 0:
    count += 1
else:
    while n > 0:
        n //= 10
        count += 1
print (count)
```

### Bài 16:
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
chan = 0
le = 0
if n == 0:
    chan += 1
else:
    while n > 0:
        chu_so = n % 10
        if chu_so % 2 == 0:
            chan += 1
        else:
            le += 1
        n //= 10
print (f"n có số chữ số chẵn là: {chan}")
print (f"n có số chữ số lẻ là: {le}")
```

### Bài 17:

### Bài 18:
#### Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không 



```python
n = int(input())
if n < 0:
    print (f"{n} khong co dang 3^k ")
else:
    temp = n
    while temp % 3 == 0:
        temp //= 3
        
    if temp == 1:
        print (f"{n} co dang 3^k ")
    else:
        print (f"{n} khong co dang 3^k ")
```

### Bài 20:
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input())

count = 0
temp = n
if n == 0:
    print("Số 0 không có chữ số nguyên tố nào.")
else:
    while temp > 0:
        chu_so = temp % 10 
        
        if chu_so in [2, 3, 5, 7]:
            count += 1
            
        temp //= 10 

    print(f"Số {n} có {count} chữ số là số nguyên tố.")
```
