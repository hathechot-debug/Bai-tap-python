### Bài 2
####  Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 


```python
a = float(input())
b = float(input())
r = a/2
dientichhcn = a*b
pi = 3.14
dientichhinhtron = pi*r*r
s = dientichhcn -dientichhinhtron
print (f"{s:.2f}")
```

```python
c = input()
if c.isalpha:
    print (f"{c} la ki tu alphabet")
else:
    print (f"{c} khong phai la ki tu alphabet")
```

### Bài 5
#### Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 


```python
c = input()
c_thuong = c.lower()
if c == "A":
    ket_qua = chr(ord(c_thuong) - 1)
    print (f"{ket_qua}")
else:
    ket_qua = chr(ord(c_thuong) - 1)
    print (f"chữ cái thường liền trước chữ {c_thuong} là : {ket_qua}")
```

### Bài 6
#### Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 


```python
a, b, c = map(int, input().split())
import math
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print (s)
else:
    print (" khong phai 3 canh cua tam giac")
```

- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng 


```python
ten = input()
chi_so_cu = int(input())
chi_so_moi = int(input())
kwh = chi_so_moi - chi_so_cu
if kwh <= 50:
    tien_dien = kwh * 1984
elif kwh <= 100:
    tien_dien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien_dien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460
tong_tien = round(tien_dien * 1.08)
print (f"Ho va ten: {ten}")
print (f"Tien phai tra la: {tong_tien}")
```
