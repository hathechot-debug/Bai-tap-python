### W6A1

```python
numbers = list(map(int, input().split()))

result = []
for num in numbers:
    if num not in result:
        result.append(num)
print(result)
```

### W6A2

```python
n = list(map(int, input().split()))
result = []
tong = 0
for num in n:
    tong += num
    result.append(tong)
print(result)
```
