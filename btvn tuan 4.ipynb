### W4A20
#### Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
if n < 0:
    print (f"{n} khong co dang 2^k ")
else:
    temp = n
    while temp % 2 == 0:
        temp //= 2
        
    if temp == 1:
        print (f"{n} co dang 2^k ")
    else:
        print (f"{n} khong co dang 2^k ")
```

### W4A21
#### Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input())
tong = 0
while n > 0:
    chu_so = n % 10
    tong += chu_so
    n //= 10
print (tong)
```

### W4A22
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
if n == 0:
    chan += 1
else:
    while n > 0:
        chu_so = n % 10
        if chu_so % 2 == 0:
            chan += 1
        else:
            le += 1
        n //= 10
print (f"Co {chan} chu so chan va {le} chu so le")
```

#### W4A25
#### Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max_value = 0
min_value = 0
while True:
    n = int(input())
    if n == -1:
        break
    if max_value is None:
        max_value = n
        min_value = n
    else:
        if n > max_value:
            max_value = n
        if n < min_value:
            min_value = n

if max_value is not None:
    print (f"Gia tri lon nhat la: {max_value}")
    print (f"Gia tri nho nhat la: {min_value}")
```

### W4A26
#### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
a = int(input())
c = 1
d = 1 
for i in range (1, a // 2 + 1):
    c, d = d, c + d
    if d >= a:
        break
print (c)
    
```
