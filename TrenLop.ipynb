```python
class Cylinder:
    def __init__(self, bankinh, chieucao):
        self.bk = bankinh
        self.cc = chieucao
    def getArea(self):
        return 2 * 3.14 * self.bk**2 + 2 * 3.14 * self.bk * self.cc
    def getVolume(self):
        return 3.14 * self.bk**2 * self.cc
    
    def __str__(self):
        return f"({self.bk}, {self.cc})"
bankinh = float(input())
chieucao = float(input())
cylinder1 = Cylinder(bankinh, chieucao)
print (cylinder1)
print(f"Dien tich be mat: {cylinder1.getArea():.2f}")
print(f"The tich: {cylinder1.getVolume():.2f}")
```

#### Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không và phương thức tính ngày kế tiếp của ngày tháng năm nào đó nếu nó là hợp lệ
#### Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
####  Đầu vào
#### Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng dd/mm/yyyy.
#### Đầu ra
#### In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng dd/mm/yyyy. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class MyDate:
    def __init__(self, d, m, y):
        self.day = d
        self.month = m
        self.year = y

    def checkValid(self):
        if self.year < 0:
            return False
        if self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > 31:
            return False
        if self.month in [4, 6, 9, 11] and self.day > 30:
            return False
        if self.month == 2:
            if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
                if self.day > 29:
                    return False
            else:
                if self.day > 28:
                    return False
        return True
    
    def getNextDay(self):
        nextDay = MyDate(self.day + 1, self.month, self.year)
        if nextDay.checkValid():
            return nextDay
        nextDay.day = 1
        nextDay.month += 1
        if nextDay.checkValid():
            return nextDay
        nextDay.month = 1
        nextDay.year += 1
        return nextDay
    
    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}" # 02d để đảm bảo ngày và tháng luôn có 2 chữ số

d, m, y = map(int, input().split("/"))
mydate1 = MyDate(d, m, y)
print(mydate1.checkValid())
print(mydate1)
if mydate1.checkValid():
    nextDay = mydate1.getNextDay()
    print (nextDay)
else:
    print ("Ngay thang nhap vao sai roi")
```
