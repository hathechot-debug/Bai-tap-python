### W5A3
####  Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            return False
    return True

n  = int(input())
print (so_nguyen_to(n))
```

### W5A4
#### VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def so_hoan_hao(n):
    if n < 1:
        return False
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
        
n = int(input())
print(so_hoan_hao(n))
```

### W5A6
#### Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(n):
    if n < 0:
        return None
    elif n == 0:
        return 1
    else:
        ket_qua = 1
        for i in range (1, n + 1):
            ket_qua *= i
        return ket_qua
    
n = int(input())
print (giai_thua(n))
```

### W5A8
#### Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng ủa chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 

```python
def hamming_distance(x: int, y: int) -> int:
    z = x ^ y
    count = 0

    while z > 0:
        if z & 1 == 1:
            count += 1
        z = z >> 1

    return count
x, y = map(int, input().split())
print (hamming_distance(x, y))
```

### W5A10
#### Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'. Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng độ dài. 


```python
def is_isomorphic(a: str, b: str) -> bool:
    map_ab = {}
    used = set()

    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]

        if char_a in map_ab:
            if map_ab[char_a] != char_b:
                return False
        else:
            if char_b in used:
                return False

            map_ab[char_a] = char_b
            used.add(char_b)

    return True

print(is_isomorphic("abca", "zbxz"))
print(is_isomorphic("ab", "cc"))
print(is_isomorphic("egg", "add"))
print(is_isomorphic("foo", "bar"))
```

### W5A12
#### Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return "Không thể chia cho 0"

    q = dividend // divisor
    r = dividend - divisor * q

    return q, r


dividend = int(input())
divisor = int(input())

result = divide_with_remainder(dividend, divisor)

print(result)
```

### W5A14
####  Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 


```python
def compute(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return None
        return a / b
    else:
        return None
    
a, b = map(float, input().split())
op = input()
print (compute(a, b, op))
     
```

### W5A15
#### Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triangle(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return False

    sides = sorted([a, b, c])
    x, y, z = sides

    if x + y <= z:
        return False

    return x * x + y * y == z * z

a, b, c = map(int, input().split())
print (is_right_triangle(a, b, c))
```
