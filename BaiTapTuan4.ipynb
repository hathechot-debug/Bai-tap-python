```python
while True:
    n = int(input())
    if n > 0:
        break

def is_prime(n):
    if n <= 1:
        return False
    for i in range (2, int(n ** 0.5)+ 1):
        if n % i == 0:
            return False
    return True
print (is_prime(n))
```

### W4A6: 
#### Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."

```python
a = int(input())
b = int(input())
tong = 0
def is_prime(n):
    if n <= 1:
        return False
    for i in range (2, int(n ** 0.5)+ 1):
        if n % i == 0:
            return False
    return True
for i in range (a, b + 1):
    if is_prime(i):
        tong += i
print(tong)
```

### W4A7:
#### Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 


```python
n = int(input())
uoc_lon_nhat = 0
i = 2
while i * i <= n:
    if n % i == 0:
        uoc_lon_nhat = i
        n //= i
    else:
        i += 1
if n > 1:
    uoc_lon_nhat = n
print(uoc_lon_nhat)
```

### W4A8
#### Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n_str = input("Nhập n: ")
steps = 0

while n_str != n_str[::-1]:
    n = int(n_str) + int(n_str[::-1])
    n_str = str(n)
    steps += 1

print(f"{steps}")
print(f"{n_str}")
```

### W4A14
#### Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n. 


```python
m, n = map(int, input().split())
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
print(gcd(m, n))
```
