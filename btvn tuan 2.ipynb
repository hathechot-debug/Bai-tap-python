### Bai 5
#### Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a = float(input())
b = float(input())
c = float(input())
if a == 0:
    if b == 0:
        if c == 0:
            print("phương trình có vô số nghiệm")
        else:
            print("phương trình vô nghiệm")
else:
    delta = b ** 2 - 4 * a * c
    print (delta)
    if delta < 0:
        print ("phương tình vô nghiệm thực")
    elif delta == 0:
        x = -b / (2 * a)
        print (f"phương trình có nghiệm kép: x1 = x2 = {x}")
    else:
        x1 = (-b + math.sqrt(delta) / (2 * a))
        x2 = (-b - math.sqrt(delta) / (2 * a))
        print ("phương trình có 2 nghiệm phân biệt:")
        print (x1)
        print (x2)
```

```python
a, b, c, d, m, n = map(float, input().split())
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
if D != 0:
    x = Dx / D
    y = Dy / D
    print (f"phuong trinh co 2 nghiem x = {x} va y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print ("phuong trinh co vo so nghiem")
    else:
        print ("phuong trinh vo nghiem")
```
