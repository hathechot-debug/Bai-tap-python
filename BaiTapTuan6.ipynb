#### Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

### WW6A3
#### Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)


```python
a = tuple(map(int, input().split()))
k = int(input())

n = len(a)

if n == 0:
    print(a)
else:
    k = k % n
    result = a[k:] + a[:k]
    print(result)
```

### W6A4
#### Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó 

```python
line = input()
pairs = line.split()
result = {}

for p in pairs:
    key, value = p.split(':')
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
print(result)
```

### W6A5
#### Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

```python
so = list(map(int, input().split()))
result = {'positive': 0, 'negative': 0, 'zero': 0}

for n in so:
    if n > 0:
        result['positive'] += 1
    elif n < 0:
        result['negative'] += 1
    else:
        result['zero'] += 1

print(result)
```

### W6A9
#### Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key. 

```python
a = input().split()
b = input().split()

d1 = {}
d2 = {}

for i in range(0, len(a), 2):
    key = a[i]
    value = int(a[i + 1])
    d1[key] = value

for i in range(0, len(b), 2):
    key = b[i]
    value = int(b[i + 1])
    d2[key] = value

new_d = {}

for key in d1:
    new_d[key] = d1[key]

for key in d2:
    if key in new_d:
        new_d[key] += d2[key]
    else:
        new_d[key] = d2[key]

result = {}

for key in sorted(new_d):
    result[key] = new_d[key]

print(result)
```

### W6A10
#### Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple

```python
a = list(map(int, input().split()))
k = int(input())

result = []

for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == k:
            pair = (a[i], a[j])
            if pair not in result:
                result.append(pair)

result.sort(key=lambda x: x[0]) # Sắp xếp theo phần tử đầu tiên của cặp

print(result)
```

### W6A12
#### Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.

```python
n = list(map(int, input().split()))

count = {}

for i in n:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1

max_count = max(count.values())

ans = None

for i in count:
    if count[i] == max_count:
        if ans is None or i < ans:
            ans = i

print(ans)
```

### W6A13
#### Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới. 

```python
a = input().split()

d = {}

for i in range(0, len(a), 2):
    key = a[i]
    value = a[i + 1]
    d[value] = key

print(d)
```

### W6A14
#### Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các phần tử chung đó, không trùng lặp. 

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))

result = []

for x in a:
    if x in b and x not in result:
        result.append(x)

print(result)
```
