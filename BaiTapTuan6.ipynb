```python
n = list(map(int, input().split()))
result = []
tong = 0
for num in n:
    tong += num
    result.append(tong)
print(result)
```

```python
line = input()
pairs = line.split()
result = {}

for p in pairs:
    key, value = p.split(':')
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
print(result)
```

```python
so = list(map(int, input().split()))
result = {'positive': 0, 'negative': 0, 'zero': 0}

for n in so:
    if n > 0:
        result['positive'] += 1
    elif n < 0:
        result['negative'] += 1
    else:
        result['zero'] += 1

print(result)
```

```python
import ast

def nhap_dict(ten_dict):
    # Nhập nguyên một chuỗi như: {"a": 1, "b": 2}
    chuoi = input(f"Nhập {ten_dict} (định dạng {{'key': value}}): ")
    # Chuyển chuỗi thành dictionary thật
    return ast.literal_eval(chuoi)

# Các bước gộp và sắp xếp bên dưới giữ nguyên

dict1 = nhap_dict("Dict 1")
dict2 = nhap_dict("Dict 2")
ket_qua = dict1.copy()

for key, value in dict2.items():
    if key in ket_qua:
        ket_qua[key] += value 
    else:
        ket_qua[key] = value  

keys_da_sap_xep = sorted(ket_qua.keys())
dict_cuoi_cung = {k: ket_qua[k] for k in keys_da_sap_xep}

print(dict_cuoi_cung)
```
