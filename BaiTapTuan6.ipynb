### W6A1
#### Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
numbers = list(map(int, input().split()))

result = []
for num in numbers:
    if num not in result:
        result.append(num)
print(result)
```

### W6A2
#### hập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự


```python
n = list(map(int, input().split()))
result = []
tong = 0
for num in n:
    tong += num
    result.append(tong)
print(result)
```

```python
line = input()
pairs = line.split()
result = {}

for p in pairs:
    key, value = p.split(':')
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
print(result)
```

### W6A5
#### Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

```python
so = list(map(int, input().split()))
result = {'positive': 0, 'negative': 0, 'zero': 0}

for n in so:
    if n > 0:
        result['positive'] += 1
    elif n < 0:
        result['negative'] += 1
    else:
        result['zero'] += 1

print(result)
```

### W6A8
#### Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary. 

```python
chuoi = input().split()
counts = {}
for word in chuoi:
    if word in counts:
        counts[word] += 1
    else:
        counts[word] = 1
print(counts)
```
