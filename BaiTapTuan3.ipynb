### W3A16
#### Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả cách nhau một dấu cách 


```python
x = float(input())

n = int(x)

# Làm tròn xuống
if x >= 0:
    tron_xuong = n
else:
    if x == n:
        tron_xuong = n
    else:
        tron_xuong = n - 1

# Làm tròn lên
if x >= 0:
    if x == n:
        tron_len = n
    else:
        tron_len = n + 1
else:
    tron_len = n

# Làm tròn tới số nguyên gần nhất
if x >= 0:
    if x - n >= 0.5:
        gan_nhat = n + 1
    else:
        gan_nhat = n
else:
    if n - x >= 0.5:
        gan_nhat = n - 1
    else:
        gan_nhat = n

print(tron_len, tron_xuong, gan_nhat)
```

### W3A17
#### Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không? Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs d. 


```python
a, b, c, d = map(int, input().split())
if a == 0:
    if b == 0 and c == 0 and d == 0:
        print ("YES")
    else:
        print ("NO")
else:
    if b % a == 0:
        q = b // a
        if b * q == c and c * q == d:
            print ("YES")
        else:
            print ("NO")
```
